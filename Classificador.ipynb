````python
# Funcao que converte a estrutura de dados
````

````python
def Converte_Dados(gesture_df, max_range):
    array = []
    inicial = 0
    final = 21    
    for i in range(max_range):
        array.append([*list(gesture_df[inicial:final].x.values),
                      *list(gesture_df[inicial:final].y.values),
                      *list(gesture_df[inicial:final].z.values),
                      *list(gesture_df[inicial:final].gesture.unique())])
        inicial = inicial + 21
        final = final + 21

    df_ok = pd.DataFrame(array)
    
    return df_ok
````

````python
Converte_Dados(gestures_ok, 10)
````

````python
# convertendo todas os gestos mapeados
````

````python
list_df_for_machine_learning = []
for i, gesture in enumerate(gestures_array):
    print(i)
    list_df_for_machine_learning.append(Converte_Dados(gesture, 100))
````

````python
df_for_machine_learning = pd.concat(list_df_for_machine_learning)
````

````python
from pycaret.classification import setup
from pycaret.classification import compare_models
# define the location of the dataset

# set column names as the column number
n_cols = df.shape[1]
df.columns = [str(i) for i in range(n_cols)]
# setup the dataset
grid = setup(data=df, target=df.columns[-1], html=True, silent=False, verbose=False, remove_perfect_collinearity=False)
# evaluate models and compare models
best = compare_models()
# report the best model
print(best)
````
